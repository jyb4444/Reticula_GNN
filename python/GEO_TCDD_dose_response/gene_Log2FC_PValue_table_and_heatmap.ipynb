```python
%%R
library(dplyr)
library(tidyr)
library(stringr)

export_file <- "/Users/kejiyuan/Desktop/export.csv"
sorted_by_ig_file <- "/Users/kejiyuan/Desktop/test/ig/sorted_by_ig0.csv"

export_df <- read.csv(export_file, stringsAsFactors = FALSE)
export_df$Pathway <- str_trim(export_df$Pathway)
export_df$ReactionLikeEvent <- str_trim(export_df$ReactionLikeEvent)

sorted_by_ig_df <- read.csv(sorted_by_ig_file, sep="\t", stringsAsFactors = FALSE, header = TRUE)
sorted_by_ig_df$start <- str_trim(sorted_by_ig_df$start)
sorted_by_ig_df$end <- str_trim(sorted_by_ig_df$end)
sorted_by_ig_df$ig_1 <- as.numeric(sorted_by_ig_df$ig_1)

filtered_df <- sorted_by_ig_df %>% filter(ig_1 > 0.5)

result_rows <- list()
for (i in 1:nrow(filtered_df)) {
  start <- filtered_df$start[i]
  end <- filtered_df$end[i]
  matched_rows <- export_df %>% filter(str_detect(ReactionLikeEvent, fixed(start)) | str_detect(ReactionLikeEvent, fixed(end)))
  result_rows[[i]] <- matched_rows 
}

result_df <- bind_rows(result_rows) %>% distinct()

print(result_df)
```

```python
%%R
library('DESeq2')
results_list <- list()
doses <- c("0.01", "0.03", "0.1", "0.3", "1", "3", "10", "30") 
for (dose in doses) {
    res <- results(dds, contrast=c("Doses", dose, "0"))
    res_shrink <- lfcShrink(dds, contrast=c("Doses", dose, "0"), res=res, type="normal")
    results_list[[dose]] <- res_shrink
}
print(results_list)
for (dose in names(results_list)) {
    print(results_list[[dose]])
    write.table(as.data.frame(results_list[[dose]]), 
                file=paste0("gene_log2fc_pval_0_vs_", dose, ".txt"),
                sep="\t",     
                row.names=FALSE,  
                quote=FALSE)
}
```

```python
%%R
library(gplots)
library(dplyr)

# Create a new column combining Gene and MGI.symbol
merged_data <- merged_data %>%
  mutate(gene_MGI = paste(Gene, "-", MGI.symbol))

# Reorder columns to place gene_MGI at the front
merged_data <- merged_data %>%
  select(gene_MGI, everything())

# Get the unique Reactions
unique_reactions <- unique(merged_data$Reactions)

# Calculate the global min and max values for the heatmaps
all_values <- unlist(merged_data[, -c(1:6)], use.names = FALSE)
global_min <- min(as.numeric(all_values), na.rm = TRUE)
global_max <- max(as.numeric(all_values), na.rm = TRUE)
color_key_range <- c(global_min, global_max)

# Function to create and save a heatmap for each reaction
create_heatmap <- function(reaction) {
  subset_data <- merged_data %>% filter(Reactions == reaction)
  
  # Debugging: print dimensions of subset data
  print(paste("Processing Reaction:", reaction))
  print(paste("Number of genes:", nrow(subset_data)))
  
  heatmap_data <- subset_data[, -c(1:7)]  # Remove non-numeric columns
  heatmap_data <- as.matrix(sapply(heatmap_data, as.numeric))  # Ensure all data is numeric
  
  # Debugging: check dimensions
  print(paste("Heatmap data dimensions:", dim(heatmap_data)))
  print(paste("Number of row names:", length(subset_data$gene_MGI)))
  
  if (nrow(heatmap_data) == length(subset_data$gene_MGI)) {
    rownames(heatmap_data) <- subset_data$gene_MGI  # Set row names to the new concatenated column
    
    # Set output file
    output_file <- paste0("/Users/kejiyuan/Desktop/test/log2fc/heatmaps/heatmap_", gsub(" ", "_", reaction), ".png")
    
    # Open graphic device
    png(file=output_file, width = 2000, height = 1600, res = 300)
    
    # Create heatmap without dendrograms
    heatmap.2(heatmap_data,
              trace = "none",
              dendrogram = "none",  # No dendrograms
              Rowv = FALSE,  # Do not reorder rows
              Colv = FALSE,  # Do not reorder columns
              col = bluered(100),  # Color palette
              scale = "none",  # Do not scale
              zlim = color_key_range,  # Fix the color key range
              breaks = seq(color_key_range[1], color_key_range[2], length.out = 101), # Ensure the color breaks are consistent
              margins = c(10, 10),  # Margins for the heatmap
              key = TRUE,  # Show color key
              keysize = 1.5,
              density.info = "none",  # No density plot
              cexRow = 0.7,  # Font size for row names
              cexCol = 0.7,  # Font size for column names
              lmat = rbind(c(4, 3), c(2, 1)),  # Layout of the plot
              lhei = c(1, 4),  # Heights of the rows in the layout
              lwid = c(1.5, 4)  # Widths of the columns in the layout
    )
    
    # Close graphic device
    dev.off()
    
    # Check if the file was successfully saved
    if (file.exists(output_file)) {
        message("File successfully saved at ", output_file)
    } else {
        warning("Failed to save the file. Please check the file path and permissions.")
    }
  } else {
    warning("Number of rows in heatmap_data does not match length of gene_MGI for reaction:", reaction)
  }
}

# Loop through each unique reaction and create a heatmap
for (reaction in unique_reactions) {
  create_heatmap(reaction)
}
```
