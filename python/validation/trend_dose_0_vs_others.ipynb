```python
import matplotlib.pyplot as plt

data = {
    "0.01 vs 0": 84,
    "0.03 vs 0": 84,
    "0.1 vs 0": 85,
    "0.3 vs 0": 85,
    "1 vs 0": 94,
    "3 vs 0": 94,
    "10 vs 0": 100,
    "30 vs 0": 100
}

x = list(data.keys())
y = list(data.values())

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Accuracy (%)")
plt.title("Accuracy vs Dose", fontsize=14)
plt.xlabel("Dose Comparison", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.savefig("accuracy_vs_dose_comparison.pdf", dpi=300, bbox_inches='tight', format='pdf')

plt.show()
```
