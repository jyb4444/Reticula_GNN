```python
%%R
library(magrittr)
OUT_DIR <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/GEO_model_training/output_summation/"
IN_DIR <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/GEO_model_training/input/"
IN_DIR2 <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/GEO_model_training/input_summation/"

X <- readRDS(paste(OUT_DIR, "rxn_pca_nls.Rds", sep = ""))
Y <- readRDS(paste(OUT_DIR,"GEO_model_training_tissue_detail_vec_train.Rds",sep=""))
E <- read.table(paste(IN_DIR,"ReactionNetwork_Rel.txt",sep=""))
```

```python
%%R
rxn2nodeLabel.nls <- list()
nodeLabel2rxn.nls <- list()
for(i in 1:length(X)){
  rxn2nodeLabel.nls[[names(X)[i]]] <- i
  nodeLabel2rxn.nls[[i]] <- names(X)[i]
}
```

```python
%%R
E <- E %>%
  dplyr::filter(V1 %in% names(rxn2nodeLabel.nls))
head(E)
```

```python
%%R
E <- E %>%
  dplyr::filter(V3 %in% names(rxn2nodeLabel.nls))
head(E)
```

```python
%%R
E <- E %>%
  dplyr::filter(V1 %in% names(rxn2nodeLabel.nls)) %>%
  dplyr::filter(V3 %in% names(rxn2nodeLabel.nls)) %>%
  dplyr::select(V1,V3)
```

```python
%%R
write.table(E,
            file=paste(IN_DIR2,"edgeLabels.csv",sep=""),
            row.names = FALSE,
            col.names = FALSE)
```

```python
%%R
node1 <- numeric()
node2 <- numeric()
for(i in 1:nrow(E)){
  node1 <- c(node1,rxn2nodeLabel.nls[[as.character(E$V1[i])]])
  node2 <- c(node2,rxn2nodeLabel.nls[[as.character(E$V3[i])]])
}
```

```python
%%R
z <- unlist(rxn2nodeLabel.nls)
y <- unlist(nodeLabel2rxn.nls)
assertthat::are_equal(length(z),length(y))
head(z)
```

```python
%%R
write.table(z,
            file=paste(IN_DIR2,"rxn2nodeLabel_nls.csv",sep=""),
            row.names = TRUE,
            col.names = FALSE)
write.table(y,
            file=paste(IN_DIR2,"nodeLabel2rxn_nls.csv",sep=""),
            row.names = TRUE,
            col.names = FALSE)
```

```python
%%R
E <- data.frame(node1 = node1,
                node2 = node2)

X <- as.data.frame(X)

Y <- as.data.frame(Y)

print(E)
```
