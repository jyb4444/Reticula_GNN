```python
%%R

plot_roc_data <- function (roc_res) 
{
    n_method <- length(unique(roc_res$Methods))
    n_group <- length(unique(roc_res$Groups))
    roc_res_df <- data.frame(Specificity = numeric(0), Sensitivity = numeric(0), 
        Group = character(0), AUC = numeric(0), Method = character(0), info = character(0))
    
    for (i in 1:n_method) {
        for (j in 1:n_group) {
            temp_data_1 <- data.frame(Specificity = roc_res$Specificity[[i]][j], 
                Sensitivity = roc_res$Sensitivity[[i]][j], Group = unique(roc_res$Groups)[j], 
                AUC = roc_res$AUC[[i]][j], Method = unique(roc_res$Methods)[i],
                info = roc_res$Specificity[[i]][16])  
            colnames(temp_data_1) <- c("Specificity", "Sensitivity", 
                "Group", "AUC", "Method", "info")
            roc_res_df <- rbind(roc_res_df, temp_data_1)
        }
        temp_data_2 <- data.frame(Specificity = roc_res$Specificity[[i]][n_group + 
            1], Sensitivity = roc_res$Sensitivity[[i]][n_group + 
            1], Group = "Macro", AUC = roc_res$AUC[[i]][n_group + 
            1], Method = unique(roc_res$Methods)[i])
        temp_data_3 <- data.frame(Specificity = roc_res$Specificity[[i]][n_group + 
            2], Sensitivity = roc_res$Sensitivity[[i]][n_group + 
            2], Group = "Micro", AUC = roc_res$AUC[[i]][n_group + 
            2], Method = unique(roc_res$Methods)[i])
        colnames(temp_data_2) <- c("Specificity", "Sensitivity", 
            "Group", "AUC", "Method")
        colnames(temp_data_3) <- c("Specificity", "Sensitivity", 
            "Group", "AUC", "Method")

        temp_data_2$info <- "empty"
        temp_data_3$info <- "empty"
        roc_res_df <- rbind(roc_res_df, temp_data_2)
        roc_res_df <- rbind(roc_res_df, temp_data_3)
    }
    return(roc_res_df)
}
plot_roc_data_df <- plot_roc_data(roc_data) %>% dplyr::filter(Group != "Micro" & Group != "Macro")
print(plot_roc_data_df)
```

```python
%%R
library(ggplot2)
library(plotly)

# Create the ggplot2 plot
figure <- ggplot(plot_roc_data_df, aes(x = 1 - Specificity, y = Sensitivity)) +
#   geom_path(aes(color = Group, linetype = Method), size = 1.5, show.legend=TRUE) +
  geom_point(aes(color = Group, shape = Method, text = paste("Info: ", info, " + Method: ", Method))) +
  annotate("segment", x = 0, xend = 1, y = 0, yend = 1, 
           colour = 'grey', linetype = 'dotdash') +
  theme_bw() + 
  ggtitle("GNN vs Resnet MultiROC")

# Convert ggplot2 plot to plotly
plotly_figure <- ggplotly(figure)

for(i in 1:length(plotly_figure$x$data)) {
  trace_name <- plotly_figure$x$data[[i]]$name
  if (grepl("gnn", trace_name)) {
    plotly_figure$x$data[[i]]$visible <- TRUE
  } else if (grepl("resnet", trace_name)) {
    plotly_figure$x$data[[i]]$visible <- TRUE
  } else {
    plotly_figure$x$data[[i]]$visible <- TRUE
  }
}

plotly_figure <- plotly_figure %>%
  layout(
    updatemenus = list(
      list(
        buttons = list(
          list(method = "restyle",
               args = list("visible", list(TRUE, FALSE)),
               label = "gnn"),
          list(method = "restyle",
               args = list("visible", list(FALSE, TRUE)),
               label = "resnet"),
          list(method = "restyle",
               args = list("visible", list(TRUE, TRUE)),
               label = "All")
        ),
        direction = "down",
        x = 1.0,
        y = 0.1,
        showactive = TRUE
      )
    )
  )

# Save plotly plot as an HTML file
htmlwidgets::saveWidget(plotly_figure, "figure.html", selfcontained = FALSE)

# Print a message to inform the user that the plot has been saved
cat("Plotly figure saved as 'figure.html'\n")
```
