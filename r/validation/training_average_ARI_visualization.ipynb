```python
import pandas as pd
import os
path = "/Users/kejiyuan/Desktop/materials/Training_ARI/"
file_names = [
    "top_10_Adipose_rxns.csv",
    "top_10_Adrenal_rxns.csv",
    "top_10_Aorta_rxns.csv",
    "top_10_Bone_rxns.csv",
    "top_10_Brain_rxns.csv",
    "top_10_Eye_rxns.csv",
    "top_10_Heart_rxns.csv",
    "top_10_Intestine_rxns.csv",
    "top_10_Kidney_rxns.csv",
    "top_10_Liver_rxns.csv",
    "top_10_Lung_rxns.csv",
    "top_10_Mammary gland_rxns.csv",
    "top_10_MOE_rxns.csv",
    "top_10_Muscle_rxns.csv",
    "top_10_Nerve_rxns.csv",
    "top_10_Olfactory_rxns.csv",
    "top_10_Ovary_rxns.csv",
    "top_10_Pancreas_rxns.csv",
    "top_10_Skin_rxns.csv",
    "top_10_Sperm_rxns.csv",
    "top_10_Spinal cord_rxns.csv",
    "top_10_Stomach_rxns.csv",
    "top_10_Tendon_rxns.csv",
    "top_10_Testes_rxns.csv",
    "top_10_Uterus_rxns.csv",
    "top_10_VNO_rxns.csv",
    ]
file_paths = [path + name for name in file_names]

data = {}
for file in file_paths:
    tissue = file.split('_')[3]
    if os.path.exists(file):
        data[tissue] = pd.read_csv(file)
    else:
        print(f"File not found: {file}")

print(data)
```

```python
ari_levels = {tissue: df['ARI'].mean() for tissue, df in data.items()}
print(ari_levels)
```

```python
import matplotlib.pyplot as plt

# Tissues and their corresponding ARI levels
tissues = list(ari_levels.keys())
ari_values = list(ari_levels.values())

# Creating the bar chart
plt.figure(figsize=(20, 6))
plt.bar(tissues, ari_values, color='skyblue')
plt.xlabel('Tissue')
plt.ylabel('Top_10 Average ARI Level')
plt.title('Top_10 Average ARI Levels Across Tissues')
plt.xticks(rotation=45)

plt.savefig("/Users/kejiyuan/Desktop/materials/Training_ARI/training_average_ari_levels.png")

plt.show()
```
