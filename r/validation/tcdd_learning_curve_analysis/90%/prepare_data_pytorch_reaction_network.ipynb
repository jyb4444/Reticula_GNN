```python
%%R
library(magrittr)
IN_DIR <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/tcdd/learning_curve_analysis/input/"
OUT_DIR <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/tcdd/learning_curve_analysis/output/"

X <- readRDS(paste(OUT_DIR, "rxn_pca_nls030_time_course_90.Rds", sep = ""))
Y <- readRDS(paste(OUT_DIR,"tcdd_dose_detail_vec_train030_time_course_90.Rds",sep=""))
E <- read.table(paste(IN_DIR,"ReactionNetwork_Rel.txt",sep=""))
```

```python
%%R
search_rmmus <- c("R-MMU-2161795", "R-MMU-2161814", "R-MMU-2161890", 
                  "R-MMU-2161899", "R-MMU-2161940", "R-MMU-9027043", 
                  "R-MMU-5423678", "R-MMU-76373", "R-MMU-76386", 
                  "R-MMU-76426", "R-MMU-9037761")
found_rmmus <- search_rmmus %in% names(X)

result_df <- data.frame(RMMU = search_rmmus, Exists = found_rmmus)

print(result_df)
```

```python
%%R
rxn2nodeLabel.nls <- list()
nodeLabel2rxn.nls <- list()
for(i in 1:length(X)){
  rxn2nodeLabel.nls[[names(X)[i]]] <- i
  nodeLabel2rxn.nls[[i]] <- names(X)[i]
}
```

```python
%%R
E <- E %>%
  dplyr::filter(V1 %in% names(rxn2nodeLabel.nls))
head(E)
```

```python
%%R
E <- E %>%
  dplyr::filter(V3 %in% names(rxn2nodeLabel.nls))
head(E)
```

```python
%%R
E <- E %>%
  dplyr::filter(V1 %in% names(rxn2nodeLabel.nls)) %>%
  dplyr::filter(V3 %in% names(rxn2nodeLabel.nls)) %>%
  dplyr::select(V1,V3)
```

```python
%%R
write.table(E,
            file=paste(IN_DIR,"edgeLabels_time_course_90.csv",sep=""),
            row.names = FALSE,
            col.names = FALSE)
```

```python
%%R
node1 <- numeric()
node2 <- numeric()
for(i in 1:nrow(E)){
  node1 <- c(node1,rxn2nodeLabel.nls[[as.character(E$V1[i])]])
  node2 <- c(node2,rxn2nodeLabel.nls[[as.character(E$V3[i])]])
}
```

```python
%%R
z <- unlist(rxn2nodeLabel.nls)
y <- unlist(nodeLabel2rxn.nls)
assertthat::are_equal(length(z),length(y))
head(z)
```

```python
%%R
write.table(z,
            file=paste(IN_DIR,"rxn2nodeLabel_nls_time_course_90.csv",sep=""),
            row.names = TRUE,
            col.names = FALSE)
write.table(y,
            file=paste(IN_DIR,"nodeLabel2rxn_nls_time_course_90.csv",sep=""),
            row.names = TRUE,
            col.names = FALSE)
```

```python
%%R
E <- data.frame(node1 = node1,
                node2 = node2)

X <- as.data.frame(X)

Y <- as.data.frame(Y)

print(E)
```
