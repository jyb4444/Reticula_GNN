```python
import pandas as pd
import matplotlib.pyplot as plt

path = "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/gtex/"

gnn_data = pd.read_csv(path + 'GNN.csv')
resnet_data = pd.read_csv(path + 'Resnet.csv')
rewired_network_data = pd.read_csv(path + 'rewired_network.csv')
shuffled_feature_data = pd.read_csv(path + 'shuffled_feature.csv')
shuffled_targets_data = pd.read_csv(path + 'shuffled_targets.csv')
summation_data = pd.read_csv(path + 'Summation.csv')

plt.figure(figsize=(10, 7))

plt.plot(resnet_data['Epoch'], resnet_data['Accuracy'], label='Resnet CV (0.99)')
plt.plot(gnn_data['Epoch'], gnn_data['Accuracy'], label='GNN (0.93)')
plt.plot(rewired_network_data['Epoch'], rewired_network_data['Accuracy'], label='Rewired Network (0.8357)')
plt.plot(summation_data['Epoch'], summation_data['Accuracy'], label='Summation (0.7361)')
plt.plot(shuffled_feature_data['Epoch'], shuffled_feature_data['Accuracy'], label='Shuffled Features (0.2775)')
plt.plot(shuffled_targets_data['Epoch'], shuffled_targets_data['Accuracy'], label='Shuffled Targets (0.2791)')

plt.xlabel('Epoch',fontsize=14)
plt.ylabel('Accuracy',fontsize=14)
plt.title('Accuracy Across Epochs for Various Networks')

plt.legend(loc='lower right', bbox_to_anchor=(0.95, 0.15), fontsize=10, frameon=True, framealpha=0.9, edgecolor='gray')

plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)

plt.tight_layout()

# Save as PDF
plt.savefig('accuracy_plot.pdf', format='pdf', dpi=300)

plt.show()
```
