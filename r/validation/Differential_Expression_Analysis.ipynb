```python
%%R
library(DESeq2)
library(ggplot2)
library(pheatmap)

OUT_DIR <- "/Users/kejiyuan/Desktop/reticula/data/tcdd/output/"

dds <- readRDS(paste0(OUT_DIR, "dds030_time_cross.Rds"))
print(dds)
```

```python
%%R
dds <- DESeq(dds)
print(dds)
```

```python
%%R
resultsNames(dds)
```

```python
%%R

res <- results(dds, contrast = list("Dose"))
print(res)
```

```python
%%R
print(dim(res))
```

```python
%%R
tail(res[order(res$log2FoldChange), ], n=20)
```

```python
%%R
rxn_df <- readRDS(paste0(OUT_DIR, "rxn2ensembls_nls030_time_cross.Rds"))
print(rxn_df)
```

```python
%%R
library(dplyr)
library(tidyr)
library(tibble) 

res_df <- as.data.frame(res)
res_df <- rownames_to_column(res_df, var = "row")  

rxn_expanded <- rxn_df %>% 
  unnest(Gene)

res_with_reactions <- res_df %>%
  left_join(rxn_expanded, by = c("row" = "Gene"))

res_with_reactions <- res_with_reactions %>% distinct()

print(tail(res_with_reactions[order(res_with_reactions$log2FoldChange), ], n=20))

write.table(res_with_reactions, file=paste0(OUT_DIR, "differential_expression_030_time_cross.txt"), sep="\t", row.names=FALSE)
```

```python
%%R
print(tail(res_with_reactions[order(res_with_reactions$log2FoldChange, decreasing = TRUE), ], n = 20))
```

```python
%%R
library(dplyr)

filtered_res <- res_with_reactions %>%
  filter(Reaction %in% c("R-MMU-6799246", "R-MMU-2993780", "R-MMU-5693609"))

print(filtered_res)
```
