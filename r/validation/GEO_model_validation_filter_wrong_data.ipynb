```python
%%R

IN_DIR <- '/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/GEO_model_validation/output/'

gnn_data <- read.csv(paste0(IN_DIR, 'gnn_predictions.tsv'), sep = "\t", row.names = NULL)
print(length(gnn_data$issues))
wrong_prediction <- gnn_data[gnn_data$target != gnn_data$prediction, ]
print(length(wrong_prediction$issues))
print(wrong_prediction)
```

```python
%%R

IN_DIR2 <- "/mnt/home/yuankeji/RanceLab/reticula_new/reticula/data/"

all_data <- read.csv(paste0(IN_DIR2, 'CleanedSamplesGNN_v3.csv'), sep=",", row.names = NULL)

head(all_data)
```

```python
%%R
print(colnames(all_data))
result <- data.frame(Project=character(), Sample=character(), Tissues=character(), Target=numeric(), Prediction=numeric(), stringsAsFactors=FALSE)
col_names <- c('Project', 'Sample', 'Tissues', 'Target', 'Prediction')
colnames(result) <- col_names

for (i in 1:nrow(wrong_prediction)) {
    pre_sample_id <- wrong_prediction[i, "X..sample_id"]
    for (j in 1:nrow(all_data)) {
        sample_id <- all_data[j, "SRR"]
        project_id <- all_data[j, "project_id"]
        if (pre_sample_id == sample_id) {
            issues <- wrong_prediction[i, 'issues']
            target <- wrong_prediction[i, 'target']
            prediction <- wrong_prediction[i, 'prediction']
            new_row <- data.frame(Project=project_id, Sample=pre_sample_id, Tissues=issues, Target=target, Prediction=prediction, stringsAsFactors=FALSE)
            result <- rbind(result, new_row)
        }
    }
}

print(result)

```

```python
%%R
write.csv(result, file="/mnt/home/yuankeji/RanceLab/reticula_new/reticula/src/r/validation/filter_wrong_data.csv")
```

```python
%%R
print(colnames(all_data))
result_all <- data.frame(Project=character(), Sample=character(), Tissues=character(), Target=numeric(), Prediction=numeric(), stringsAsFactors=FALSE)
col_names <- c('Project', 'Sample', 'Tissues', 'Target', 'Prediction')
colnames(result_all) <- col_names

for (i in 1:nrow(gnn_data)) {
    pre_sample_id <- gnn_data[i, "X..sample_id"]
    for (j in 1:nrow(all_data)) {
        sample_id <- all_data[j, "SRR"]
        project_id <- all_data[j, "project_id"]
        if (pre_sample_id == sample_id) {
            issues <- gnn_data[i, 'issues']
            target <- gnn_data[i, 'target']
            prediction <- gnn_data[i, 'prediction']
            new_row <- data.frame(Project=project_id, Sample=pre_sample_id, Tissues=issues, Target=target, Prediction=prediction, stringsAsFactors=FALSE)
            result_all <- rbind(result_all, new_row)
        }
    }
}

print(result_all)
```

```python
%%R
library(dplyr)
result_summary <- result_all %>%
  mutate(Correct = Target == Prediction) %>%
  group_by(Tissues) %>%
  summarise(
    Total_Predictions = n(),
    Correct_Predictions = sum(Correct),
    Correct_Rate = Correct_Predictions / Total_Predictions,
    Incorrect_Predictions = Total_Predictions - Correct_Predictions,
    Incorrect_Rate = Incorrect_Predictions / Total_Predictions
  )
print(result_summary)
```

```python
%%R
print(colnames(result_summary))
write.csv(result_summary, file="/mnt/home/yuankeji/RanceLab/reticula_new/reticula/src/r/validation/summary.csv")
```
